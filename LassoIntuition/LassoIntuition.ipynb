```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nx = 151
ny = 101
x = np.linspace(-10, 10, nx)
y = np.linspace(-10, 10, ny)
X, Y = np.meshgrid(x, y)
```

```python
# elliptical bowl
xc = 6
yc = 3
Z = 2*(X-xc)**2 + .5*(X-xc)*(Y-yc) + 1*(Y-yc)**2
CS = plt.contour(X, Y, Z, 20)
plt.show()
```

```python
#1-norm by itself
Z1 = abs(X) + abs(Y)
CS1 = plt.contour(X, Y, Z1, 20)
```

```python
# sum with lambda=1
lamda = 1
Zlasso = Z + lamda*Z1
CSlasso = plt.contour(X, Y, Zlasso, 20)
```

```python
# sum with lambda=5
lamda = 5
Zlasso = Z + lamda*Z1
CSlasso = plt.contour(X, Y, Zlasso, 40)
```
